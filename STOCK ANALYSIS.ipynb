```python
#Combined analysis
#Change in stock price over time  From 2018 to 2024
allstocks=['AAPL','GOOG','MSFT','AMZN']
all_stocks_data=yf.download(allstocks,start="2022-01-01",end="2024-01-01")
plt.figure(figsize=(20,9))
for i in allstocks:
    plt.plot(all_stocks_data['Close'][i],label=i,linewidth=3)

plt.title("Change in stock price over time  From 2018 to 2024",fontsize=20)
plt.xlabel("Date",fontsize=30)
plt.ylabel("Closing price",fontsize=35) 
plt.legend(loc=7,fontsize=30,bbox_to_anchor=[1.17,0.6],edgecolor='black')
plt.grid()
```

```python
#Combined analysis
#Change in stock Volume Trade over time  From 2018 to 2024
allstocks=['AAPL','GOOG','MSFT','AMZN']
all_stocks_data=yf.download(allstocks,start="2022-01-01",end="2024-01-01")
plt.figure(figsize=(20,9))
for i in allstocks:
    plt.plot(all_stocks_data['Volume'][i],label=i,linewidth=3)

plt.title("Change in stock Volume Trade over time  From 2018 to 2024",fontsize=20)
plt.xlabel("Date",fontsize=30)
plt.ylabel("Volume Trade",fontsize=35) 
plt.legend(loc=7,fontsize=30,bbox_to_anchor=[1.17,0.6],edgecolor='black')
plt.grid()
```

```python
#Moveing average various Stocks
plt.figure(figsize=(20,9))
for i in allstocks:
    plt.plot(all_stocks_data['Close'],linewidth=3)
    plt.plot(all_stocks_data['Close'][i].rolling(window=50).mean(),label=i,linewidth=3)

plt.title("Moveing average various Stocks From 2018 to 2024",fontsize=20)
plt.xlabel("Date",fontsize=30)
plt.ylabel("Price",fontsize=35) 
plt.legend(loc=7,fontsize=30,bbox_to_anchor=[1.17,0.6],edgecolor='black')
plt.grid()
```

```python
#Daily Return average of a stock
plt.figure(figsize=(15,4))
for i in allstocks:
    
    all_stocks_data['Adj Close'][i].pct_change().plot(label=i,kind='kde',legend=True)



plt.title("Daily Returns of the Stock From 2018 to 2024",fontsize=20)
plt.xlabel("Density",fontsize=20)
plt.ylabel("Daily Returns",fontsize=20)
plt.grid()
```

```python
#Daily return of a average stock
daily_return = all_stocks_data['Adj Close'].pct_change()
daily_return_avg = daily_return.mean()
print('Average Daily Return:\n')
print(daily_return_avg)
```

```python
daily_return
```

```python
# Add a new column ‘Trend’ whose values are based on the 'Daily Return'
all_stocks_data['Trend'] = ''
for stock in allstocks:
    all_stocks_data.loc[all_stocks_data['Adj Close'][stock].pct_change() > 0, 'Trend'] = 'Up'
    all_stocks_data.loc[all_stocks_data['Adj Close'][stock].pct_change() < 0, 'Trend'] = 'Down'
    all_stocks_data.loc[all_stocks_data['Trend'] == '', 'Trend']='Stable'
```

```python
all_stocks_data
```

```python
# Visualize trend frequency through a Pie Chart
plt.figure(figsize=(10, 10))
plt.pie(all_stocks_data['Trend'].value_counts(), labels=all_stocks_data['Trend'].value_counts().index, autopct='%1.1f%%',startangle=90,rotatelabels=True,labeldistance=1.05,textprops={'fontsize':20})
plt.title('Trend Frequency',loc='left',fontdict={'fontsize':30})
plt.legend(title='Trend Frequency of Stocks',bbox_to_anchor=(1.6,1),fontsize=40)
plt.show()
```
